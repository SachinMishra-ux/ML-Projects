## I am going to deal the NA Values in the following manner.
### 1) since location NA value is 1. I will directly drop the NA for location
### 2) For size,bath and balcony I will replace NA values with median.

```python
df1
```

```python
df1.dropna(subset=['location'])# Droped NA Values from 'location' column
```

```python
df1['bath'].fillna(int(df1['bath'].median()), inplace=True)# Replaced NA values of 'bath' Column by median
df1['balcony'].fillna(int(df1['balcony'].median()), inplace=True) # Replaced NA values of 'balcony' Column by median
```

### Now let's work on total_sqft column

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df2[~df2['total_sqft'].apply(is_float)]
```

## Clearly observe the 'total_sqft' column. It has lot's of crazy values and I need to handle all of them. so let's do this.

## Above shows that total_sqft can be a range (e.g. 2100-2850). For such case we can just take average of min and max value in the range. There are other cases such as 34.46Sq. Meter which one can convert to square ft using unit conversion.

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        pass   
```

```python
df3 = df2.copy()
df3.total_sqft = df3.total_sqft.apply(convert_sqft_to_num)
df3
```
